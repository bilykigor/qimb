```python
w=60
f = lambda x:x[0]*x[2]+x[1]*x[3]
def GetFeatures(driverID,j):
    driverDir = '/home/user1/Desktop/SharedFolder/Kaggle/FeaturesCleaned/Cos/'+str(driverID)
    tripFiles = range(1,201)
   
    cos = pd.DataFrame(np.zeros((200,51)))
    for index,tripID in enumerate(tripFiles):       
        #print index,tripID
        trip = Trip(driverID,tripID,pd.read_csv(driverDir+'_' + str(tripID) + '.csv'))
        trip.coordinates=trip.coordinates[(trip.coordinates.x!=0) | (trip.coordinates.y!=0)]
        trip.coordinates=trip.coordinates.reset_index(drop=True)
        if trip.coordinates.shape[0]<w+50:
            continue
        
        df=trip.coordinates.ix[w:,:].copy()
        df=df.reset_index(drop=True)
        df['x2']=trip.coordinates.ix[:trip.coordinates.shape[0]-w,:].x
        df['y2']=trip.coordinates.ix[:trip.coordinates.shape[0]-w,:].y
        
        tmp = np.asarray(trip.Quantiles(df.apply(f,axis=1)))
        #print tripID,len(tmp)
        cos.loc[index] = np.asarray(trip.Quantiles(df.apply(f,axis=1))) 
    
    cos.to_csv('/home/user1/Desktop/SharedFolder/Kaggle/FeaturesCleaned/CosQuantiles/' + str(driverID) + '.csv', index=False)
    #del cos

    return 0
```

```python
import time
from joblib import Parallel, delayed  
import multiprocessing
num_cores=multiprocessing.cpu_count()

'''for driverID in driversFiles:
    print driverID
    MakePrediction(driverID,1)
'''
num_cores=4
inputs = range(len(driversFiles)/num_cores)
for i in inputs:
    start = time.time()
    r=range(num_cores*i,num_cores*i+num_cores)
    Parallel(n_jobs=num_cores)(delayed(MakePrediction)(driversFiles[j],j % num_cores) for j in r) 
    #print time.time()-start 
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.ensemble import GradientBoostingClassifier as GBC
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.neighbors import KNeighborsClassifier as KNN
import EnsembleClassifier as eclf
reload(eclf)

randTrips = range(200)
drivers_sampleSize=5
trips_sampleSize=150

def MakePrediction(cur_driver,tk):
    #print cur_driver
    Xpred = pd.read_csv('/home/user1/Desktop/SharedFolder/Kaggle/FeaturesCleaned/CosQuantiles/' + str(cur_driver) + '.csv')
    #Xpred=Xpred.fillna(0, inplace=True)
    cur_driver_df = pd.DataFrame(np.zeros((Xpred.shape[0],3)),columns=['driver','trip','prob'])
    cur_driver_df.driver = cur_driver
    cur_driver_df.trip = range(1,201)
    
    np.random.shuffle(randTrips)
    Xtrain = Xpred.loc[randTrips[:trips_sampleSize]]
    ytrain = pd.DataFrame(np.ones((trips_sampleSize,1)))
    ytrain.ix[:] = int(1)
    
    np.random.shuffle(randomDrivers)           
    for rand_driver in randomDrivers[:drivers_sampleSize]:
        if rand_driver==cur_driver:
            continue
        
        D = pd.read_csv('/home/user1/Desktop/SharedFolder/Kaggle/FeaturesCleaned/CosQuantiles/' + str(rand_driver) + '.csv')
        #D=D.fillna(0, inplace=True)
        np.random.shuffle(randTrips)
        Xtrain = Xtrain.append(D.loc[randTrips[:trips_sampleSize]])
        tmp = pd.DataFrame(np.zeros((trips_sampleSize,1)))
        tmp.ix[:] = int(0)
        ytrain = ytrain.append(tmp)  
        
    Xtrain=Xtrain.reset_index(drop=True)
    ytrain=ytrain.reset_index(drop=True)
    
    #preprocessing
    #pca.fit(Xtrain)
    
    
    #fit model  
    #===============================================================================
    clf = \
    eclf.EnsembleClassifier(
    clfs=[
    LR(class_weight='auto',C=0.5)
    ,RF()
    ,GBC()
    ])

    tmpInd = pd.isnull(Xtrain).any(1).nonzero()[0]
    if (len(tmpInd)>0):
        print Xtrain.loc[tmpInd]
    clf.fit(Xtrain,np.asarray(ytrain[0]))
    #===============================================================================
    cur_driver_df.prob=clf.predict_proba(Xpred)[:,1]
    
    cur_driver_df.to_csv('/home/user1/Desktop/SharedFolder/Kaggle/results/' + str(cur_driver) + '.csv', index=False)  
    
    return 0
```
